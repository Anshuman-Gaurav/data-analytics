```python
import pandas as pd
import numpy as np

# ---------------------------------------------------------
# 1. Load and Inspect
# ---------------------------------------------------------
# Replace 'supply_chain_data.csv' with your actual file name
df = pd.read_csv('APL_Logistics.csv', encoding='utf-8')

# Display basic information to check data types and missing values
print("--- Initial Data Info ---")
print(df.info())

# ---------------------------------------------------------
# 2. Financial Validation & Currency Normalization
# ---------------------------------------------------------
# Sometimes financial data loads as strings with currency symbols (e.g., "$1,200.50"). 
# If your data is already purely numeric, this step acts as a safety net.
financial_cols = [
    'Sales', 'Order Profit Per Order', 'Benefit per order', 'Sales per customer',
    'Order Item Discount', 'Order Item Product Price', 'Order Item Total', 'Product Price'
]

# Function to clean currency symbols and convert to float
def clean_currency(x):
    if isinstance(x, str):
        x = x.replace('$', '').replace(',', '')
    return pd.to_numeric(x, errors='coerce')

for col in financial_cols:
    if col in df.columns:
        df[col] = df[col].apply(clean_currency)

# ---------------------------------------------------------
# 3. Remove Inconsistent or Zero-Value Records
# ---------------------------------------------------------
# Drop rows where critical identifying information is missing
critical_id_cols = ['Customer Id', 'Order Id', 'Category Id', 'Product Name']
df.dropna(subset=[col for col in critical_id_cols if col in df.columns], inplace=True)

# Remove zero-value or illogical records (e.g., negative quantity or zero sales)
# We keep negative profit, as loss-making orders are crucial for your analysis
df = df[(df['Sales'] > 0) & (df['Order Item Quantity'] > 0)]

# ---------------------------------------------------------
# 4. Validate Sales and Profit Fields (Derived KPIs)
# ---------------------------------------------------------
# Create a standardized Profit Margin (%) column to ensure consistency across your analysis
df['Calculated_Profit_Margin_%'] = (df['Order Profit Per Order'] / df['Sales']) * 100

# Validate 'Order Item Total' to ensure it aligns with the math: 
# (Price * Quantity) - Discount
df['Expected_Order_Total'] = (df['Order Item Product Price'] * df['Order Item Quantity']) - df['Order Item Discount']

# Create a validation flag to see if the provided total matches our expected total (allowing for minor rounding differences)
df['Total_Validation_Check'] = np.isclose(df['Order Item Total'], df['Expected_Order_Total'], atol=0.05)

# Optional: Print out how many rows failed the financial math check
failed_checks = len(df[df['Total_Validation_Check'] == False])
if failed_checks > 0:
    print(f"Warning: {failed_checks} rows have mismatched Order Item Totals compared to Price * Quantity - Discount.")

# ---------------------------------------------------------
# 5. Final Output Check
# ---------------------------------------------------------
print("\n--- Cleaned Data Shape ---")
print(df.shape)

print("\n--- Summary Statistics of Financials ---")
print(df[['Sales', 'Order Profit Per Order', 'Calculated_Profit_Margin_%']].describe())
```

```python
import pandas as pd
import numpy as np

# ---------------------------------------------------------
# 1. Load and Inspect
# ---------------------------------------------------------
# Replace 'supply_chain_data.csv' with your actual file name
df = pd.read_csv('APL_Logistics.csv', encoding='latin1')

# Display basic information to check data types and missing values
print("--- Initial Data Info ---")
print(df.info())

# ---------------------------------------------------------
# 2. Financial Validation & Currency Normalization
# ---------------------------------------------------------
# Sometimes financial data loads as strings with currency symbols (e.g., "$1,200.50"). 
# If your data is already purely numeric, this step acts as a safety net.
financial_cols = [
    'Sales', 'Order Profit Per Order', 'Benefit per order', 'Sales per customer',
    'Order Item Discount', 'Order Item Product Price', 'Order Item Total', 'Product Price'
]

# Function to clean currency symbols and convert to float
def clean_currency(x):
    if isinstance(x, str):
        x = x.replace('$', '').replace(',', '')
    return pd.to_numeric(x, errors='coerce')

for col in financial_cols:
    if col in df.columns:
        df[col] = df[col].apply(clean_currency)

# ---------------------------------------------------------
# 3. Remove Inconsistent or Zero-Value Records
# ---------------------------------------------------------
# Drop rows where critical identifying information is missing
critical_id_cols = ['Customer Id', 'Order Id', 'Category Id', 'Product Name']
df.dropna(subset=[col for col in critical_id_cols if col in df.columns], inplace=True)

# Remove zero-value or illogical records (e.g., negative quantity or zero sales)
# We keep negative profit, as loss-making orders are crucial for your analysis
df = df[(df['Sales'] > 0) & (df['Order Item Quantity'] > 0)]

# ---------------------------------------------------------
# 4. Validate Sales and Profit Fields (Derived KPIs)
# ---------------------------------------------------------
# Create a standardized Profit Margin (%) column to ensure consistency across your analysis
df['Calculated_Profit_Margin_%'] = (df['Order Profit Per Order'] / df['Sales']) * 100

# Validate 'Order Item Total' to ensure it aligns with the math: 
# (Price * Quantity) - Discount
df['Expected_Order_Total'] = (df['Order Item Product Price'] * df['Order Item Quantity']) - df['Order Item Discount']

# Create a validation flag to see if the provided total matches our expected total (allowing for minor rounding differences)
df['Total_Validation_Check'] = np.isclose(df['Order Item Total'], df['Expected_Order_Total'], atol=0.05)

# Optional: Print out how many rows failed the financial math check
failed_checks = len(df[df['Total_Validation_Check'] == False])
if failed_checks > 0:
    print(f"Warning: {failed_checks} rows have mismatched Order Item Totals compared to Price * Quantity - Discount.")

# ---------------------------------------------------------
# 5. Final Output Check
# ---------------------------------------------------------
print("\n--- Cleaned Data Shape ---")
print(df.shape)

print("\n--- Summary Statistics of Financials ---")
print(df[['Sales', 'Order Profit Per Order', 'Calculated_Profit_Margin_%']].describe())
```

```python
import plotly.express as px

print("--- 1. High-Level KPIs ---")
total_revenue = df['Sales'].sum()
total_profit = df['Order Profit Per Order'].sum()
overall_margin = (total_profit / total_revenue) * 100

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Overall Profit Margin: {overall_margin:.2f}%\n")

# ---------------------------------------------------------
# 2. Product & Category Profitability Analysis
# ---------------------------------------------------------
# Group data by category to see which ones make or lose money
category_perf = df.groupby('Category Name').agg({
    'Sales': 'sum',
    'Order Profit Per Order': 'sum'
}).reset_index()

# Calculate margin per category
category_perf['Category Margin (%)'] = (category_perf['Order Profit Per Order'] / category_perf['Sales']) * 100

# Sort by profit so the chart looks organized
category_perf = category_perf.sort_values(by='Order Profit Per Order', ascending=False)

# Create an interactive bar chart
fig_cat = px.bar(
    category_perf, 
    x='Category Name', 
    y='Order Profit Per Order', 
    color='Category Margin (%)',
    title="Profitability by Category (Color-coded by Margin)",
    color_continuous_scale='RdYlGn', # Red (bad) to Green (good)
    height=600
)
fig_cat.show()

# ---------------------------------------------------------
# 3. Discount Impact Diagnostics (Margin Erosion)
# ---------------------------------------------------------
# Group by exact discount rates to see the average profit margin at each level
discount_impact = df.groupby('Order Item Discount Rate').agg({
    'Calculated_Profit_Margin_%': 'mean'
}).reset_index()

# Create a line chart to find the "Danger Zone"
fig_disc = px.line(
    discount_impact, 
    x='Order Item Discount Rate', 
    y='Calculated_Profit_Margin_%',
    title="Discount-Driven Margin Erosion: When do discounts cause losses?",
    markers=True
)
# Add a red dashed line at the 0% profit mark (Break-Even Point)
fig_disc.add_hline(y=0, line_dash="dash", line_color="red", annotation_text="Break-Even (0% Profit)")
fig_disc.show()
```
